### 5. Coefficient of Determination R^2

#### reference:
#### https://scikit-learn.org/stable/modules/generated/sklearn.metrics.r2_score.html
#### https://en.wikipedia.org/wiki/Coefficient_of_determination
#### https://en.wikipedia.org/wiki/Pearson_correlation_coefficient

```python
from sklearn.metrics import r2_score
import numpy as np
import matplotlib.pyplot as plt
# from scipy.stats import pearsonr
```

```python
# given target
y_true = [5, 10, 11, 16, 19]

# base line
y_mean = [np.mean(y_true) for i in range(len(y_true))]
```

```python
# fit a line
from sklearn import linear_model
X = [1, 2, 3, 4, 5]
X = np.asarray(X).reshape(-1, 1)
Y = y_true
model = linear_model.LinearRegression()
model.fit(X, Y)

print(model.intercept_)
print(model.coef_[0])
```

```python
# regression line
y_pred = [3.4*i+2 for i in range(1,(len(y_true))+1)]
```

```python
# calculate using formula
var_mean = sum([(y_true[i]-y_mean[i])**2 for i in range(len(y_true))])
var_pred = sum([(y_true[i]-y_pred[i])**2 for i in range(len(y_true))])
r2 = (var_mean-var_pred)/(var_mean)
print(r2)
```

```python
# calculate using scikit learn
r2_score(y_true, y_pred)
```

```python
# calculate using pearson correlation
correlation = np.corrcoef(y_true, y_pred)[0][-1]
correlation**2
```

```python
plt.style.use('ggplot')

plt.plot(y_true,
         color='red', marker='o', linestyle='dashed',
         linewidth=1, markersize=7)
plt.plot(y_pred,
         color='green', marker='o', linestyle='solid',
         linewidth=1, markersize=7)

plt.plot(y_mean,
         color='blue', marker='o', linestyle='dashed',
         linewidth=0.5, markersize=5)

plt.title('Plot')
plt.xlabel('ID')
plt.ylabel('Target')
plt.show()
```
