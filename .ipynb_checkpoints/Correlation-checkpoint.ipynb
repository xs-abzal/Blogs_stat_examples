### Read in data.

```python
# https://www.kaggle.com/hmavrodiev/london-bike-sharing-dataset?select=london_merged.csv
df = pd.read_csv('data/london_merged.csv')
df.head(3).append(df.tail(3))
```

```python
data = df.iloc[:,2:].copy()
data.head()
```

```python
# covariance
data.cov()
```

```python
# correlation
data.corr()
```

```python
abs(data.corr()) > 0.70
```

```python
# t1 and t2 are strongly correlated
```

```python
### PCA example
from sklearn.decomposition import PCA
```

```python
pca = PCA(random_state = 42)
pca.fit(data)
pca.n_components_
```

```python
plt.style.use('ggplot')

plt.plot(np.cumsum(pca.explained_variance_ratio_),
         color='blue', marker='o', linestyle='dashed',
         linewidth=1, markersize=7)

plt.title('Cumulative Explained Variance')
plt.xlabel('Components number')
plt.ylabel('Explained Variance in %')
plt.show()
```

```python
# ref to eigen vector 
# http://www.physics.utah.edu/~detar/lessons/python/numpy_eigen/node1.html
# https://machinelearningmastery.com/calculate-principal-component-analysis-scratch-python/
# https://towardsdatascience.com/let-us-understand-the-correlation-matrix-and-covariance-matrix-d42e6b643c22
# https://towardsdatascience.com/understanding-pca-fae3e243731d

# get eigen values and eigen vectors
```

```python
# ex.corr() 
# np.corrcoef(ex.T)
```

```python
from sklearn.preprocessing import StandardScaler
ex_scaled = StandardScaler().fit_transform(ex)
ex_scaled
```

```python
# np.cov(ex_scaled.T) or below the same
pd.DataFrame(ex_scaled).cov()
```

```python
e_val, e_vect = np.linalg.eig(pd.DataFrame(ex_scaled).cov()) 
```

```python
# quick check
np.dot(pd.DataFrame(ex_scaled).cov(),e_vect)
#np.dot(np.cov(ex_scaled.T),e_vect) or in numpy
```

```python
# quick check
e_val*e_vect
```

```python
e_val
```

```python
e_vect.T.dot(ex_scaled.T).T
```

# the same with PCA but ordered per eigen values

```python
pca = PCA()
pca.fit(ex_scaled)
pca.transform(ex_scaled)
```

```python
pca.explained_variance_ratio_
```

```python
plt.style.use('ggplot')

plt.plot(pca.explained_variance_ratio_,
         color='blue', marker='o', linestyle='dashed',
         linewidth=1, markersize=7)

plt.plot(np.cumsum(pca.explained_variance_ratio_),
         color='blue', marker='o', linestyle='dashed',
         linewidth=1, markersize=7)

plt.title('Cumulative Explained Variance')
plt.xlabel('Components number')
plt.ylabel('Explained Variance in %')
plt.show()
```
